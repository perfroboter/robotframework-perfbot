```python
import pandas as pd
import matplotlib.pyplot as plt

testresults = pd.read_csv("database-export.csv")

testresults
```

```python
boxplot = testresults.boxplot(column=['elapsedTime'], by=['longname'], rot=45, fontsize=6)
```

```python
valid_login_df = testresults.loc[testresults['longname'] == 'Testcases.Example-Tests.Valid Login.Valid Login']
valid_login_df
```

```python
plt.scatter(10000,'Testcases.Example-Tests.Valid Login.Valid Login',marker="o")
boxplot = valid_login_df.boxplot(column=['elapsedTime'], by=['longname'], vert=False)
boxplot.scatter(10000,'Testcases.Example-Tests.Valid Login.Valid Login',marker="x")
```
