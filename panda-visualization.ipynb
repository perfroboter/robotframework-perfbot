```python
import pandas as pd
import matplotlib.pyplot as plt

testresults = pd.read_csv("database-export.csv")

testresults
```

```python
boxplot = testresults.boxplot(column=['elapsedTime'], by=['longname'], rot=45, fontsize=6)
```

```python
valid_login_df = testresults.loc[testresults['longname'] == 'Testcases.Example-Tests.Valid Login.Valid Login']
valid_login_df
```

```python
boxplot = valid_login_df.boxplot(column=['elapsedTime'], by=['longname'], rot=0)
```

```python
other_df = testresults.loc[testresults['longname'] != 'Testcases.Example-Tests.Valid Login.Valid Login']
other_df
```

```python
boxplot = other_df.boxplot(column=['elapsedTime'], by=['longname'], rot=90, fontsize=10)
```
